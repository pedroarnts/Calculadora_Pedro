```python
print("***CALCULADORA***")

continuar = True

while continuar:

  #Valores
  n1 = float(input("Digite o 1º valor: "))
  n2 = float(input("Digite o 2º valor: "))

  #Opções
  print("\nEscolha um Operador:")
  print("1 - Soma")
  print("2 - Subtração")
  print("3 - Multiplicação")
  print("4 - Divisão")
  print("5 - Sair")

  opção = input("Digite o número da opção desejada: ")

  #Estrutura Condicional
  if opção == "1":
    resultado = n1 + n2
    print(f"\nResultado: {n1} + {n2} = {resultado}")

  elif opção == "2":
    resultado = n1 - n2
    print(f"\nResultado: {n1} - {n2} = {resultado}")

  elif opção == "3":
    resultado = n1 * n2
    print(f"\nResultado: {n1} x {n2} = {resultado}")

  elif opção == "4":
    if n2 == 0:
      print("\nInválido. Não é possível dividir por 0.")
    else:
      resultado = n1 / n2
      print(f"\nResultado: {n1} ÷ {n2} = {resultado}")

  elif opção == "5":
    print("\nAté logo!")
    continuar = False

  else:
    print("\nOpção inválida! Digite um número de 1 a 5.")

  if continuar:
    print("\n--------------------------")
```
